```python
df.fillna(df.median(),inplace = True)
```

```python
# after handling null values

df.isnull().sum()
```

### Handling outlier using IQR


```python
for column in df.select_dtypes(exclude='object'):
    q1 = df[column].quantile(0.25)
    q3 = df[column].quantile(0.75)
    iqr = q3 - q1
    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr
    lower_outliers = (df[column] < lower)
    upper_outliers = (df[column] > upper)
    df.loc[lower_outliers, column] = q1
    df.loc[upper_outliers, column] = q3
```

### Split data

```python
X = df.drop('Life expectancy', axis =1)
y = df['Life expectancy']
```

```python
X.shape  ,y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train ,X_test , y_train , y_test = train_test_split(X,y,test_size = 0.3,random_state =42)
```

```python
numeric_columns = X_train.select_dtypes(include=['int64', 'float64']).columns
X_train_numeric = X_train[numeric_columns]
X_test_numeric = X_test[numeric_columns]
```

### Standrilaze  data

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_numeric)
X_test_scaled = scaler.transform(X_test_numeric)
```

### linear Regression with CV

```python
from sklearn.model_selection import cross_val_score,cross_val_predict
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
LR = LinearRegression()
scores = cross_val_score(LR,X_train_scaled,y_train,cv=5)
scores.mean()
```

```python
LR.fit(X_train_scaled ,y_train)
y_train_pred = LR.predict(X_train_scaled)
y_pred = LR.predict(X_test_scaled)
```

### EValuate LR

```python
from sklearn.metrics import mean_squared_error
# calculate RMSE
rmse_train = mean_squared_error(y_train ,y_train_pred,squared = False)
rmse_test = mean_squared_error(y_test ,y_pred,squared = False)
print('RMSE for training : {:.2f}'.format(rmse_train))
print('RMSE for testing :{:.2f}'.format(rmse_test))
print('Linear Regression Model Score')
print('Train Score {:.2f}'.format(LR.score(X_train_scaled,y_train)))
print('Test Score {:.2f}'.format(LR.score(X_test_scaled,y_test)))
```

```python
plt.plot(y_test ,color= 'blue',label ='Actual Value')
plt.plot(y_pred ,color = 'green' , label = 'Predicted Value')
plt.title('Actual Vs Predicted Life Expectancy')
plt.legend()
plt.show()
```

```python
plt.plot(y_train_pred ,color ='blue',label ='Train prediction')
plt.plot(y_pred ,color = 'red',label = 'Test prediction')
plt.legend()
plt.show()
```

```python
features = X.columns.values
coefficients = LR.coef_
sorted_idx = np.argsort(np.abs(coefficients))
plt.barh(range(len(sorted_idx)), coefficients[sorted_idx], align='center')
plt.yticks(range(len(sorted_idx)), [features[i] for i in sorted_idx])
plt.xlabel('Coefficient Magnitude')
plt.title('Linear Regression Feature Importance')
plt.show()
```

```python
# Assuming you have calculated RMSE values (e.g., rmse_train, rmse_test)

rmse_train_i = int(rmse_train)  

ecophs = range(1, rmse_train_i + 1) 

plt.plot(ecophs, 'bo-', label='Training RMSE')
# plt.show()
plt.plot(ecophs, 'ro-', label='Testing RMSE')
plt.xlabel('Ecophs')
plt.ylabel('RMSE')
plt.legend()
plt.show()
```

### KNN 

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import GridSearchCV
knn = KNeighborsRegressor()
k_grid = {'n_neighbors': range(1,10)}
grid_search = GridSearchCV(knn, k_grid, cv=5, scoring='neg_mean_squared_error')
grid_search.fit(X_train_scaled, y_train)
print("Best K:", grid_search.best_params_['n_neighbors'])
# Get the best model from the grid search
best_knn_model = grid_search.best_estimator_
y_pred = best_knn_model.predict(X_test_scaled)
test_mse = mean_squared_error(y_test, y_pred)
print("Test Mean Squared Error:", test_mse)
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error

knn = KNeighborsRegressor(n_neighbors =3)
knn.fit(X_train_scaled, y_train)
y_train_pred = knn.predict(X_train_scaled)
y_pred = knn.predict(X_test_scaled)
rmse_train = mean_squared_error(y_train,y_train_pred,squared= False)
rmse_test = mean_squared_error(y_test,y_pred , squared = False)
print('RMSE Train: {:.2f}',rmse_train)
print('RMSE test: {:.2f}',rmse_test)

train_score = knn.score(X_train_scaled, y_train)
test_score = knn.score(X_test_scaled, y_test)
print('Train Score {:.2f}'.format(train_score))
print('Test Score {:.2f}'.format(test_score))
```
